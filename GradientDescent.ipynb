```python
import numpy as np
import matplotlib.pyplot as plt
import math
from matplotlib import cm
from mpl_toolkits.mplot3d import Axes3D

def f(a,b):
	return a**2 - b**2

def gradient_descent(theta0, iters, alpha):
	history = [theta0] # to store all thetas
	theta = theta0     # initial values for thetas
	# main loop by iterations:
	for i in range(iters):
		# gradient is [2x, 2y]:
		gradient = [2.0*x for x in theta] #함수 f(x,y)의 미분
		# update parameters:
		theta = [a - alpha*b for a,b in zip(theta, gradient)]
		history.append(theta)
	return history

history = gradient_descent(theta0 = [4, 3], iters = 100, alpha = 0.1)

# f(x,y) = x^2 + y^2 함수의 그래프 그리기
fig = plt.figure(figsize=(20, 15))
ax = fig.gca(projection='3d')
#plt.hold(True)
a = np.arange(-4, 4, 0.1)
b = np.arange(-4, 4, 0.1)
a, b = np.meshgrid(a, b)
c = f(a,b)
surf = ax.plot_surface(a, b, c, rstride=1, cstride=1, alpha=0.2, 
		 linewidth=0, antialiased=False, cmap='inferno')

a = np.array([x[0] for x in history])
b = np.array([x[1] for x in history])
c = f(a,b)
ax.scatter(a, b, c, color="b"); 

print(c)
ax.view_init(15, 60)
plt.xlabel('x1')
plt.ylabel('x2')


plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import math
from matplotlib import cm
from mpl_toolkits.mplot3d import Axes3D

def gradient_descent(theta0, epoch, step_size):
	history = [theta0] # to store all thetas
	theta = theta0     # initial values for thetas
	# main loop by iterations:
	for i in range(epoch):
		# gradient is [x1, x2]:
		gradient = [x for x in theta] #함수 Objective(x1,x2)의 미분
		# update parameters:
		theta = [a - step_size*b for a,b in zip(theta, gradient)]
		history.append(theta)
	return history

history = gradient_descent(theta0 = [-4, 3], epoch = 100, step_size = 0.1)

# Objective(x1,x2) = x1^2 + x2^2 함수의 그래프 그리기
fig = plt.figure(figsize=(20, 15))
ax = fig.gca(projection='3d')
#plt.hold(True)
x1 = np.arange(-4, 4, 0.1)
x2 = np.arange(-4, 4, 0.1)
x1, x2 = np.meshgrid(x1, x2)
O = x1**2 - x2**2
surf = ax.plot_surface(x1, x2, O, rstride=1, cstride=1, alpha=0.2, 
		 linewidth=0, antialiased=False, cmap='inferno')

x1 = np.array([x[0] for x in history])
x2 = np.array([x[1] for x in history])
O = x1**2 - x2**2
ax.scatter(x1, x2, O, color="black"); 

print(O)
ax.view_init(15, 60)

plt.show()
```
